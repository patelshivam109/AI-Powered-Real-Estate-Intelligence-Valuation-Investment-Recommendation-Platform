```python
df_a["Sale_Date"] = (
    pd.to_datetime("1899-12-30")
    + pd.to_timedelta(df_a["Date"], unit="D")
)
```

```python
df_a[["Date", "Sale_Date"]].head()
```

```python
df_a["Sale_Year"] = df_a["Sale_Date"].dt.year
df_a["Sale_Month"] = df_a["Sale_Date"].dt.month
df_a["Sale_Quarter"] = df_a["Sale_Date"].dt.quarter
```

```python
df_a[
    ["Sale_Date", "Sale_Year", "Sale_Month", "Sale_Quarter"]
].head()
```
